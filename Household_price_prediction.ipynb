```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
labels = data['price']
conv_dates = [1 if values == 2014 else 0 for values in data.date ]
data['date'] = conv_dates
train1 = data.drop(['id', 'price'],axis=1)
```

```python
x_train,x_test,y_train,y_test = tts(train1,labels,test_size=0.1,random_state=2)
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr_pred=lr.predict(x_test)
lr_pred_train=lr.predict(x_train)
```

```python
plt.rcParams["figure.figsize"] = (10,5)
plt.subplot(1, 2, 2)
plt.scatter(lr_pred,y_test)
plt.subplot(1, 2, 1)
plt.scatter(lr_pred_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
mse_train = mean_squared_error(y_train, lr_pred_train)
mse_val = mean_squared_error(y_test, lr_pred)
# Plot the training and validation MSE
plt.plot([1, 2], [mse_train, mse_val], marker='o')
plt.xticks([1, 2], ['Training', 'Validation'])
plt.xlabel('Dataset')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Training vs. Validation Performance')
plt.grid(True)
plt.show()
```

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 500, max_depth = 6, min_samples_split = 4,learning_rate = 0.05, loss = 'ls')
```

```python
clf.fit(x_train, y_train);
clf.score(x_test,y_test)
```

```python
clf_pred_train=clf.predict(x_train)
clf_pred=clf.predict(x_test)
```

```python
plt.rcParams["figure.figsize"] = (10,5)
plt.subplot(1, 2, 1)
plt.scatter(clf_pred_train,y_train)
plt.subplot(1, 2, 2)
plt.scatter(clf_pred,y_test)
```

```python
plt.rcParams["figure.figsize"] = (10,5)
plt.subplot(1, 2, 1)
plt.scatter(lr_pred,y_test)
plt.subplot(1, 2, 2)
plt.scatter(clf_pred,y_test)
```

```python
fig = plt.figure()
sns.distplot((clf_pred-y_test), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)    
```
